```python
# TODO: Train a simple regression tree
from sklearn.tree import DecisionTreeRegressor

X = [[1], [2], [3], [4]]
y = [2.0, 4.0, 6.0, 8.0]
model = DecisionTreeRegressor(max_depth=2, random_state=42)
model.fit(X, y)
print("Prediction for x=2.5:", model.predict([[2.5]])[0])
```
