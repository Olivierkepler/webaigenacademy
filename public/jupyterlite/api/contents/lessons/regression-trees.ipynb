# Regression Trees Practice Lab

**Learning objective:** Predict a continuous target with leaf **means**, score splits with squared error, and compare numeric thresholds on a tiny height → weight style dataset.

Leaf prediction:

$$\hat{y} = \frac{1}{n}\sum y_i$$

Mean squared error:

$$MSE = \frac{1}{n}\sum (y_i - \hat{y})^2$$


```python
import numpy as np
import pandas as pd

# x = hours studied, y = score
data = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "score": [52, 55, 58, 61, 70, 74, 78, 82],
})
print(data)
```

```python
def leaf_prediction(y):
    return float(np.mean(y))

def mse(y):
    y = np.asarray(y, dtype=float)
    if len(y) == 0:
        return 0.0
    mu = y.mean()
    return float(np.mean((y - mu) ** 2))

def weighted_mse(y_left, y_right):
    n = len(y_left) + len(y_right)
    if n == 0:
        return 0.0
    return (len(y_left) / n) * mse(y_left) + (len(y_right) / n) * mse(y_right)

print("Parent mean:", leaf_prediction(data["score"]))
print("Parent MSE:", round(mse(data["score"]), 3))
```

## Guided example — one numeric split

Try `hours <= 4.5` and compare parent MSE vs weighted child MSE.


```python
t = 4.5
left = data[data["hours"] <= t]["score"].to_numpy()
right = data[data["hours"] > t]["score"].to_numpy()
print("Left mean/MSE:", round(leaf_prediction(left), 2), round(mse(left), 3))
print("Right mean/MSE:", round(leaf_prediction(right), 2), round(mse(right), 3))
print("Weighted MSE:", round(weighted_mse(left, right), 3))
```

## Exercise 1 — compare candidate thresholds

Evaluate thresholds `[2.5, 4.5, 6.5]` and pick the lowest weighted MSE.


```python
# TODO Exercise 1
thresholds = [2.5, 4.5, 6.5]
rows = []
for t in thresholds:
    left = data[data["hours"] <= t]["score"].to_numpy()
    right = data[data["hours"] > t]["score"].to_numpy()
    rows.append({"threshold": t, "weighted_mse": round(weighted_mse(left, right), 3)})
summary = pd.DataFrame(rows)
print(summary)
best_t = summary.loc[summary["weighted_mse"].idxmin(), "threshold"]
print("Best threshold:", best_t)
```

## Exercise 2 — scikit-learn check (optional but supported)

Fit `DecisionTreeRegressor(max_depth=2)` and predict for `hours=4.5`.


```python
# TODO Exercise 2
from sklearn.tree import DecisionTreeRegressor

X = data[["hours"]].to_numpy()
y = data["score"].to_numpy()
model = DecisionTreeRegressor(max_depth=2, random_state=42)
model.fit(X, y)
print("Prediction for 4.5 hours:", model.predict([[4.5]])[0])
```
