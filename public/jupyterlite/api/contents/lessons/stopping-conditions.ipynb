# Stopping Conditions Practice Lab

**Learning objective:** See how `max_depth`, `min_samples_split`, `min_samples_leaf`, pure nodes, and “no improvement” stop tree growth.

We use the Iris dataset (bundled with scikit-learn — no internet required).


```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import numpy as np

iris = load_iris()
X, y = iris.data, iris.target
print("Samples:", X.shape[0], "Features:", X.shape[1], "Classes:", len(np.unique(y)))
```

## 1) max_depth

Grow trees with `max_depth` in `{1, 3, None}` and print the actual depth and leaf count.


```python
for depth in [1, 3, None]:
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model.fit(X, y)
    print(f"max_depth={depth!r:5} -> tree_depth={model.get_depth()} leaves={model.get_n_leaves()}")
```

## 2) min_samples_split and min_samples_leaf

- `min_samples_split`: a node splits only if it has at least this many samples.
- `min_samples_leaf`: each leaf must contain at least this many samples.

Run both sweeps below and compare depth / leaf counts.


```python
print("min_samples_split sweep")
for m in [2, 20, 50]:
    model = DecisionTreeClassifier(min_samples_split=m, random_state=42)
    model.fit(X, y)
    print(f"min_samples_split={m:2d} -> depth={model.get_depth()} leaves={model.get_n_leaves()}")

print("\nmin_samples_leaf sweep")
for m in [1, 10, 30]:
    model = DecisionTreeClassifier(min_samples_leaf=m, random_state=42)
    model.fit(X, y)
    print(f"min_samples_leaf={m:2d} -> depth={model.get_depth()} leaves={model.get_n_leaves()}")
```

## Exercise 1 — pure-node stopping intuition

On a tiny pure dataset, a tree should stop immediately (depth 0 leaf).

**Expected:** depth `0`.


```python
# TODO Exercise 1
X_pure = np.array([[0.0], [0.1], [0.2]])
y_pure = np.array([1, 1, 1])
model = DecisionTreeClassifier(random_state=42)
model.fit(X_pure, y_pure)
print("Pure-node tree depth:", model.get_depth())
print("Predicted class:", model.predict([[0.15]])[0])
```

## Exercise 2 — no-improvement / constrained growth

Train one unconstrained tree and one tightly constrained tree on Iris. Compare depths and leave a short comment about overfitting risk.


```python
# TODO Exercise 2
unconstrained = DecisionTreeClassifier(random_state=42)
constrained = DecisionTreeClassifier(max_depth=2, min_samples_leaf=10, random_state=42)
unconstrained.fit(X, y)
constrained.fit(X, y)
print("Unconstrained depth/leaves:", unconstrained.get_depth(), unconstrained.get_n_leaves())
print("Constrained depth/leaves:", constrained.get_depth(), constrained.get_n_leaves())
# Reflection note: the unconstrained tree can memorize training quirks; constraints stop early.
```
