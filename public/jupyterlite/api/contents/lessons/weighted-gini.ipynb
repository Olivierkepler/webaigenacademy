# Weighted Gini Practice Lab

**Learning objective:** Score a *split* by combining left and right child Gini values, weighted by how many samples land in each child.

$$\text{Weighted Gini} = \frac{n_L}{n} G_L + \frac{n_R}{n} G_R$$

We will reuse a tiny Age / Heart Disease table, split at Age thresholds, and pick the better split.


```python
import numpy as np
import pandas as pd

patients = pd.DataFrame([
    {"age": 42, "heart_disease": 0},
    {"age": 45, "heart_disease": 0},
    {"age": 48, "heart_disease": 0},
    {"age": 50, "heart_disease": 0},
    {"age": 53, "heart_disease": 1},
    {"age": 55, "heart_disease": 1},
    {"age": 58, "heart_disease": 1},
    {"age": 60, "heart_disease": 1},
])
print(patients)
```

```python
def gini_impurity(y):
    y = np.asarray(y)
    if len(y) == 0:
        return 0.0
    counts = np.bincount(y)
    props = counts / counts.sum()
    return float(1.0 - np.sum(props ** 2))

def weighted_gini(y_left, y_right):
    n_l, n_r = len(y_left), len(y_right)
    n = n_l + n_r
    if n == 0:
        return 0.0
    return (n_l / n) * gini_impurity(y_left) + (n_r / n) * gini_impurity(y_right)

print("Helpers ready.")
```

## Guided example — split at Age <= 51.5

Compute left/right labels, each child's Gini, then weighted Gini.

**Expected:** left is mostly (or all) class 0; right is mostly class 1; weighted Gini is lower than the parent Gini.


```python
threshold = 51.5
left = patients[patients["age"] <= threshold]
right = patients[patients["age"] > threshold]

y_all = patients["heart_disease"].to_numpy()
y_left = left["heart_disease"].to_numpy()
y_right = right["heart_disease"].to_numpy()

print("Parent Gini:", round(gini_impurity(y_all), 3))
print("Left size/Gini:", len(y_left), round(gini_impurity(y_left), 3))
print("Right size/Gini:", len(y_right), round(gini_impurity(y_right), 3))
print("Weighted Gini:", round(weighted_gini(y_left, y_right), 3))
```

## Exercise 1 — another candidate

Repeat the calculation for `threshold = 46.5`.

Store `wg_46` as the weighted Gini.

**Expected:** a printed weighted Gini between 0 and 0.5.


```python
# TODO Exercise 1
threshold = 46.5
left = patients[patients["age"] <= threshold]
right = patients[patients["age"] > threshold]
y_left = left["heart_disease"].to_numpy()
y_right = right["heart_disease"].to_numpy()
wg_46 = weighted_gini(y_left, y_right)
print(f"Weighted Gini at 46.5: {wg_46:.3f}")
```

## Exercise 2 — compare candidates and pick the best

Evaluate thresholds `[46.5, 51.5, 56.5]` and select the threshold with the **lowest** weighted Gini.


```python
# TODO Exercise 2
candidates = [46.5, 51.5, 56.5]
rows = []
for t in candidates:
    left = patients[patients["age"] <= t]["heart_disease"].to_numpy()
    right = patients[patients["age"] > t]["heart_disease"].to_numpy()
    wg = weighted_gini(left, right)
    rows.append({"threshold": t, "weighted_gini": round(wg, 3), "n_left": len(left), "n_right": len(right)})

summary = pd.DataFrame(rows)
print(summary)
best = summary.loc[summary["weighted_gini"].idxmin()]
print("Best split:", dict(best))
```
