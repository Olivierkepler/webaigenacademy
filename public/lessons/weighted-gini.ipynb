Placeholder notebook for weighted Gini practice.

```python
# Threshold Splits Practice Lab
# Step 1: Create the dataset

import pandas as pd

patients = [
    {"patient": "P1",  "age": 42, "resting_bp": 118, "heart_disease": 0},
    {"patient": "P2",  "age": 45, "resting_bp": 122, "heart_disease": 0},
    {"patient": "P3",  "age": 45, "resting_bp": 128, "heart_disease": 0},
    {"patient": "P4",  "age": 48, "resting_bp": 130, "heart_disease": 0},
    {"patient": "P5",  "age": 50, "resting_bp": 135, "heart_disease": 0},
    {"patient": "P6",  "age": 53, "resting_bp": 138, "heart_disease": 1},
    {"patient": "P7",  "age": 55, "resting_bp": 142, "heart_disease": 1},
    {"patient": "P8",  "age": 55, "resting_bp": 145, "heart_disease": 1},
    {"patient": "P9",  "age": 58, "resting_bp": 148, "heart_disease": 1},
    {"patient": "P10", "age": 60, "resting_bp": 150, "heart_disease": 1},
    {"patient": "P11", "age": 63, "resting_bp": 155, "heart_disease": 1},
    {"patient": "P12", "age": 67, "resting_bp": 160, "heart_disease": 1},
]

df = pd.DataFrame(patients)

print("Medical Dataset")
display(df)
```

```python
# TODO 1:
# Extract the Age values into a list called ages.

ages = df["age"].tolist()

print(ages)
```

```python
# TODO 2:
# Remove duplicate ages and sort them.

unique_ages = sorted(set(ages))

print(unique_ages)
```

```python
# TODO 3:
# Compute candidate thresholds (midpoints).

thresholds = []

for i in range(len(unique_ages)-1):
    midpoint = (unique_ages[i] + unique_ages[i+1]) / 2
    thresholds.append(midpoint)

print("Candidate Thresholds")

for t in thresholds:
    print(t)
```

```python
# TODO 4:
# Split the patients using Age <= 51.5

threshold = 51.5

left_group = df[df["age"] <= threshold]
right_group = df[df["age"] > threshold]

print("Left Group")
display(left_group)

print()

print("Right Group")
display(right_group)
```

```python
# TODO 5 (Optional):
# Visualize the split.

import matplotlib.pyplot as plt

colors = df["heart_disease"].map({
    0: "royalblue",
    1: "crimson"
})

plt.figure(figsize=(8,6))

plt.scatter(
    df["age"],
    df["resting_bp"],
    c=colors,
    s=90
)

for _, row in df.iterrows():
    plt.text(
        row["age"]+0.2,
        row["resting_bp"]+0.5,
        row["patient"],
        fontsize=8
    )

plt.axvline(
    threshold,
    color="green",
    linestyle="--",
    linewidth=2,
    label=f"Threshold = {threshold}"
)

plt.xlabel("Age")
plt.ylabel("Resting Blood Pressure")
plt.title("Candidate Threshold Split")
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
