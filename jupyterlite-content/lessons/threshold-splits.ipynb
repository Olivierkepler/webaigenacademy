````python
# Optional visualization

import matplotlib.pyplot as plt

threshold = 51.5

no_disease = df[df["Heart Disease"] == 0]
has_disease = df[df["Heart Disease"] == 1]

fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(
    no_disease["Age"],
    no_disease["Resting BP"],
    color="blue",
    label="No Disease",
    s=80,
)
ax.scatter(
    has_disease["Age"],
    has_disease["Resting BP"],
    color="red",
    label="Disease",
    s=80,
)
ax.axvline(threshold, linestyle="--", color="black", label=f"Threshold = {threshold}")

ax.set_xlabel("Age")
ax.set_ylabel("Resting Blood Pressure")
ax.set_title("Heart Disease Patients by Age and Resting BP")
ax.legend()
plt.show()
````
