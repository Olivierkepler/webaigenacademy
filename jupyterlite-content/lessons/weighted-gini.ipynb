# Weighted Gini

Placeholder notebook for weighted Gini practice.

```python
# TODO: Compute weighted Gini for a simple split
def gini(class0, class1):
    total = class0 + class1
    if total == 0:
        return 0
    p0, p1 = class0 / total, class1 / total
    return 1 - (p0 ** 2 + p1 ** 2)

left = gini(6, 4)
right = gini(2, 8)
weighted = (10 / 20) * left + (10 / 20) * right
print(f"Weighted Gini: {weighted:.3f}")
```
